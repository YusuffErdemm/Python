```python
def fibonacci():
    a=1
    b=1
    
    yield a #` ilk olarak yield seklinde uretiyoruz
    yield b

# ve bizden istendigi surece fibonnacci serisini devam ettirelim
    while True:
        a,b=b,a+b
        yield b
```

```python
for i in fibonacci():
    if i > 100000000000000000000000000:
        break
    else:
        print(i)
        
```
