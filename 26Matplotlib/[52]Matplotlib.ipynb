```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.arange(2, 11, 2)

fig = plt.figure()  # Figure olusturduk
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.3])

# Grafiklerimizi Cizelim 
axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(x,y)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")


plt.show()
```

## Figure Ozellestirme

```python
fig,axes=plt.subplots(nrows = 2 ,ncols=1)

for ax in axes:
    ax.plot(x,y) #  Axisler uzerinde gezin ve x ` e karsilik y graph cizdir.

plt.tight_layout() # Birbirine girmesin diye.
```

```python
# Veya sadece birine cizdirmek istersek

fig,axes=plt.subplots(nrows = 2 ,ncols=1)

axes[0].plot(x,y)
axes[0].set_title("FirstAxes")
axes[1].plot(y,x)
axes[1].set_title("SecondAxes")
plt.tight_layout()
```

## Uygulama

    1. Figure Oluturun.
    2. Oluturduunuz Figure'n hepsini kaplayacak bir axes ekleyin.
    3. Aadaki grafie benzer bir grafik oluturun

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="red")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")
```

     1. Figure Size' (8,6) olan bir figure Oluturun
     2. axes1 ve axes2 adnda 2 adet axes oluturun.(axes1 pozisyonu = [0,0.1,0.8,0.8], 
     axes2 pozisyonu = [0.1,0.5,0.4,0.3])
     3. axes1 zerinde x vs z grafiini, axes2 zerinde x vs y grafiini izin ve aadaki 
     grafie benzer bir grafik oluturun. 

```python
fig=plt.figure(figsize=(8,6))
axes1=fig.add_axes([0,0.1,0.8,0.8])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])
axes1.plot(x,z,"purple")
axes2.plot(x,y,"red")
```

    1. 2 adet yanyana subplot oluturun. (figsize=(10,6))
    2. Birinci grafie (x,y) grafiini , ikinci grafie (x,z) grafiini izin.
    3. plot() fonksiyonunun parametrelerini vererek, grafii aadaki grafie benzer zelletirin.

```python
fig,axes = plt.subplots(nrows =1 ,ncols = 2,figsize=(10,6))

axes[0].plot(x,y,color ="red",lw=3,ls ="--",marker ="o",markerfacecolor ="black")
axes[0].set_xlabel("X Values")
axes[0].set_ylabel("Y Values")
axes[0].set_title("X vs Y Graph")

axes[1].plot(x,z,color ="purple",lw=3,ls = "-.")
axes[1].set_xlabel("X Values")
axes[1].set_ylabel("Z Values")
axes[1].set_title("X vs Z Graph")

```
